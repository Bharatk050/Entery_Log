```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("LogFiles.csv")
data.head()
data.info()
```

```python
data = data[data['Week'] != 'scrip']
```

```python
data.info()
```

```python
data['Time'] = pd.to_datetime(data["Time"], errors='coerce').dt.time
time_counts = data["Time"].value_counts().head(20)
print(time_counts)

# Create the bar graph
# plt.figure(figsize=(10, 6))  
# time_counts.plot(kind='bar', color='blue')
# plt.xticks(rotation=0)  
# plt.xlabel('Time')
# plt.ylabel('Count')
# plt.title('Most Frequent Times (Top 20)')
```

```python
#ERROR COUNT 
error_column = 'Event'  
error_counts = {}

for error in data[error_column]:
    if error in error_counts:
        error_counts[error] += 1
    else:
        error_counts[error] = 1
        
for error, count in error_counts.items():
  print(f"Error: {error}, Count: {count}")
```

```python
sorted_data = dict(sorted(error_counts.items(), key=lambda item: item[1], reverse=True))

top_20_data = dict(list(sorted_data.items())[:20])

labels = list(top_20_data.keys())
values = list(top_20_data.values())

plt.figure(figsize=(15, 10))
plt.bar(labels, values) 
plt.xlabel("Number of Times")
plt.ylabel("Typer of Error")
plt.title("Top 20 Error ")
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()

plt.show()
```

```python
data_count=data["EventTemplate"]
notice_count = 0
error_count = 0

for event in data_count:
    if event == "notice":
        notice_count += 1
    else:
        error_count += 1
counts=[notice_count,error_count]

x=(["Error","Notice"])
y=([error_count,notice_count])
plt.bar(x, y, color = "blue")
plt.show()

print("notice_count:-",notice_count,", error_count:-",error_count)
```

```python
#Merging 
data['date'] = data['Day'].astype(str) + '-' + data['Month'].astype(str)

#Printing Count
day_counts = data['date'].value_counts().head(20)
print(day_counts)

#Plot
plt.figure(figsize=(10, 6))  
day_counts.plot(kind='bar', color='blue')
plt.xlabel('Date')
plt.ylabel('Count')
plt.title('Day Repetitions (Top 20)')
plt.show()
```
